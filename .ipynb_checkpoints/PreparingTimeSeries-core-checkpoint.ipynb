```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn import set_config

from pmdarima.model_selection import train_test_split
from pmdarima.arima.utils import ndiffs

import statsmodels.tsa.api as tsa
```

```python
url ="https://docs.google.com/spreadsheets/d/e/2PACX-1vQN3Ksa9szQuO4G1-msXWAp17KtVHQCBnuEieu_auu1wSiBf3-krHusIx5VBMkihxj-KZLBosDIGEyR/pub?output=csv"
df = pd.read_csv(url)
df
```

```python
# Displaying all columns
all_columns = df.columns
all_columns
```

### Filter the 4 largest cities into a new dataframe.
- Tip: the "SizeRank" column has already ranked the cities by size. The larger the city, the smaller the rank value.
- Therefore the 4 largest cities would have rank values of [0,1,2,3]

```python
# Filter the 4 largest cities
largest_cities = df[df['SizeRank'].isin([0, 1, 2, 3])]
largest_cities
```

### Melt the data to long-form and prepare it for time series analysis.
- Convert the melted dates into datetime datatype.
- Make the datetime column the index.

```python
# Melt the data to long-form
melted_df = pd.melt(
    largest_cities,
    id_vars=['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', 'State', 'Metro', 'CountyName'],
    var_name='Date',
    value_name='HomeValue')
melted_df
```

```python
# Check unique values in the 'Date' column
print(melted_df['Date'].unique())
```

```python
# Convert the melted dates into datetime datatype
melted_df['Date'] = pd.to_datetime(melted_df['Date'])
melted_df.info()

#df_sun['date'] = pd.to_datetime(df_sun['date']), format='%Y-%m-%d'
#df_sun.info()
```

```python
col = melted_df.columns
col
```

```python
# Make the datetime column the index
# Set the index to the melted DataFrame
Melt_df = melted_df.set_index('Date')
Melt_df
```

```python
# Resample the dataframe as monthly frequency, grouped by City
#monthly_df = Melt_df.groupby(['SizeRank', 'RegionName', pd.Grouper(key='Date', freq='M')]).mean()
monthly_df = Melt_df.groupby(['RegionName']).resample('MS').mean(numeric_only=False)

monthly_df

```

### Once you've prepared the dataframe with the time series data for the 4 largest cities:
- Plot the home values for all 4 cities. (Hint: use unstack)
- Make sure to add a title and axis labels.
- Reformat the y-axis ticks to use thousands of dollars with a "K" at the end. (e.g. "200K, 400K, etc")


```python
# Plot the home values for all 4 cities
fig, ax = plt.subplots(figsize=(12, 8))
monthly_df['HomeValue'].unstack(level='RegionName').plot(ax=ax)

# Add title and axis labels
plt.title('Home Values for the 4 Largest Cities')
plt.xlabel('Date')
plt.ylabel('Home Value')

# Reformat the y-axis ticks to use thousands of dollars
formatter = FuncFormatter(lambda x, _: f'{int(x/1000)}K')
ax.yaxis.set_major_formatter(formatter)
```

### # Answer the questions using pandas

#### 1) Which City had the highest Typical Home Value at the end of 2008? Which had the least?


```python
# Use unstack to pivot the 'RegionName' level
end_of_2008_values_unstacked = monthly_df.unstack(level='RegionName')
end_of_2008_values_unstacked
```

```python
end_of_2008_values = end_of_2008_values_unstacked.loc['2008-12-01']
end_of_2008_values
```

```python
max_value_city = end_of_2008_values['HomeValue'].idxmax()
min_value_city = end_of_2008_values['HomeValue'].idxmin()


print(f"1) City with the highest Typical Home Value at the end of 2008: {max_value_city}")
print(f"   City with the least Typical Home Value at the end of 2008: {min_value_city}")

```

#### 2) How much did the home values change from November 2008 to December 2008 (in dollars)?


```python
november_2008 = end_of_2008_values_unstacked.loc['2008-11-01']
november_2008
```

```python
december_2008 = end_of_2008_values_unstacked.loc['2008-12-01']
december_2008
```

```python
# Calculate the change in home values
change_in_home_values = december_2008['HomeValue'] - november_2008['HomeValue']
change_in_home_values
```

```python
# Select the data for november  to December 2008
oct_to_dec_2008 = end_of_2008_values_unstacked.loc['2008-11-01':'2008-12-01']
oct_to_dec_2008
```

```python
# Calculate the change in home values using .diff()
change_in_home_values = oct_to_dec_2008['HomeValue'].diff()
change_in_home_values
```

```python
# Print the result
print(f"Home values change from October 2008 to November 2008:")
print(change_in_home_values)
```
