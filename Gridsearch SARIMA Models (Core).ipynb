### Filter the 4 largest cities into a new dataframe.
- Tip: the "SizeRank" column has already ranked the cities by size. The larger the city, the smaller the rank value.
- Therefore the 4 largest cities would have rank values of [0,1,2,3]

```python
# Filter the 4 largest cities
largest_cities = df[df['SizeRank'].isin([0, 1, 2, 3])]
largest_cities
```

### Melt the data to long-form and prepare it for time series analysis.
- Convert the melted dates into datetime datatype.
- Make the datetime column the index.

```python
# Melt the data to long-form
melted_df = pd.melt(
    largest_cities,
    id_vars=['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName', 'State', 'Metro', 'CountyName'],
    var_name='Date',
    value_name='HomeValue')
melted_df
```

```python
# Check unique values in the 'Date' column
print(melted_df['Date'].unique())
```

```python
# Convert the melted dates into datetime datatype
melted_df['Date'] = pd.to_datetime(melted_df['Date'])
melted_df.info()

#df_sun['date'] = pd.to_datetime(df_sun['date']), format='%Y-%m-%d'
#df_sun.info()
```

```python
col = melted_df.columns
col
```

```python
# Make the datetime column the index
# Set the index to the melted DataFrame
Melt_df = melted_df.set_index('Date')
Melt_df
```

```python
# Filter data to include values starting in 2018
Melt_df = Melt_df['2018':]
Melt_df
```

```python
# Check and handle null values if any
Melt_df = Melt_df.dropna()
Melt_df
```

```python
# Decompose the time series to detect seasonality
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Extract the 'HomeValue' column for New York, assuming 'New York' is in 'RegionName'
new_york_series = Melt_df[Melt_df['RegionName'] == 'New York']['HomeValue']
new_york_series
```

```python
# Extract the 'HomeValue' column for New York, assuming 'New York' is in 'RegionName'
new_york_series = Melt_df[Melt_df['RegionName'] == 'New York']['HomeValue']
```

```python
# Decompose the time series
decomposition = seasonal_decompose(new_york_series, model='additive', period=12)
```

```python

   # Plot the original time series, trend, seasonality, and residuals
plt.figure(figsize=(12, 8))

plt.subplot(411)
plt.plot(new_york_series, label='Original')
plt.legend(loc='upper left')

plt.subplot(412)
plt.plot(decomposition.trend, label='Trend')
plt.legend(loc='upper left')

plt.subplot(413)
plt.plot(decomposition.seasonal, label='Seasonality')
plt.legend(loc='upper left')

plt.subplot(414)
plt.plot(decomposition.resid, label='Residuals')
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

### Determine Time Series Model Orders (continued)

```python
# Check for stationarity
result_adf = adfuller(new_york_series)
print("ADF Statistic:", result_adf[0])
print("p-value:", result_adf[1])
print("Critical Values:", result_adf[4])
```

```python
# Perform differencing
diff_data = new_york_series.diff().dropna()
diff_data
```

```python
# Check ACF/PACF plots for orders
plt.figure(figsize=(12, 6))
plt.subplot(211)
plot_acf(diff_data, lags=20, ax=plt.gca())
plt.subplot(212)
plot_pacf(diff_data, lags=20, ax=plt.gca())
plt.show()
```

```python
# Define the size of the training set (80% of the data)
train_size = int(len(new_york_series) * 0.8)

# Split the time series into training and test sets
train, test = new_york_series[:train_size], new_york_series[train_size:]

# Display the size of the training and test sets
print("Training set size:", len(train))
print("Test set size:", len(test))
```

```python
p, d, q = 1, 1, 1  
P, D, Q, m = 1, 1, 1, 12 
```

```python
# Create and fit the SARIMA model
sarima_model = SARIMAX(train, order=(p, d, q), seasonal_order=(P, D, Q, m))
sarima_result = sarima_model.fit()
sarima_result.summary()
```

```python
# Make forecasts with the model
forecast = sarima_result.get_forecast(steps=len(test))
predicted_values = forecast.predicted_mean
predicted_values
```

```python
# Plot the forecasts versus the test data
plt.figure(figsize=(12, 6))
plt.plot(train, label='Training Data')
plt.plot(test, label='Actual Test Data')
plt.plot(predicted_values, label='SARIMA Forecast', color='red')
plt.title('Manual SARIMA Forecast vs Actual Test Data')
plt.legend()
plt.show()
```

```python
# Obtain metrics for evaluation
mse = mean_squared_error(test, predicted_values)
print("Mean Squared Error (MSE):", mse)
```

#  Tune with pmdarima's auto_arima

```python
# Use auto_arima to find the best parameters
auto_arima_model = auto_arima(train, seasonal=True, m=12, suppress_warnings=True)
best_params = auto_arima_model.get_params()
best_params
```

```python
# Print the best parameters
print("Best Parameters from auto_arima:", best_params)
```

```python
#  Fit a model on training data with the best parameters from auto_arima
best_sarima_model = SARIMAX(train, order=(best_params['order']), seasonal_order=(best_params['seasonal_order']))
best_sarima_result = best_sarima_model.fit()
best_sarima_result.summary()
```

```python
# Make forecasts with the best model
best_forecast = best_sarima_result.get_forecast(steps=len(test))
best_predicted_values = best_forecast.predicted_mean
best_predicted_values
```

```python
# Plot the forecasts versus the test data
plt.figure(figsize=(12, 6))
plt.plot(train, label='Training Data')
plt.plot(test, label='Actual Test Data')
plt.plot(best_predicted_values, label='Best SARIMA Forecast', color='green')
plt.title('Best SARIMA Forecast vs Actual Test Data')
plt.legend()
plt.show()
```

```python
# Obtain metrics for evaluation
best_mse = mean_squared_error(test, best_predicted_values)
print("Mean Squared Error (MSE) for Best SARIMA Model:", best_mse)
```

##  Make True Future Forecasts using the entire dataset

```python
# Combine the training and test sets to train the final model on the entire dataset
full_train = new_york_series[:len(train) + len(test)]
full_train
```

```python
# Fit the final model on the entire dataset
final_model = SARIMAX(full_train, order=(best_params['order']), seasonal_order=(best_params['seasonal_order']))
final_result = final_model.fit()
final_result.summary()
```

```python
# Make forecasts for the future period (6 months beyond the dataset)
future_steps = 6
final_forecast = final_result.get_forecast(steps=future_steps)
final_forecast
```

```python
# Calculate the predicted net change in home value for the forecasted period
raw_delta = final_forecast.predicted_mean[-1] - final_forecast.predicted_mean[0]
percent_change = (raw_delta / final_forecast.predicted_mean[0]) * 100
percent_change
```

```python
# Display the predicted home value and percent change
print("Predicted Home Value 6 months into the future:", final_forecast.predicted_mean[-1])
print("Percent Change in Home Values:", percent_change)
```

```python
#  Plot the True Future Forecasts
plt.figure(figsize=(12, 6))
plt.plot(full_train, label='Historical Data')
plt.plot(final_forecast.predicted_mean, label='True Future Forecast', color='purple')
plt.title('True Future Forecast vs Historical Data')
plt.legend()
plt.show()
```
